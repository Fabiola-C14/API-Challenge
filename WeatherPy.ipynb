# WeatherPy

```python
#Import pandas, requests, pprint, citipy, config key

import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import time
import scipy.stats as st
import json
import pprint
from citipy import citipy
from datetime import datetime
from config import api_key

# Create outbound file (CSV)
output_data_file = "data_city_list.csv"
```

```python
# Set coordinates
lat_coord = (-90, 90)
lng_coord = (-180, 180)

#  Create list for longitude, latitude, city and country
lat_lng = []
city_list = []
country_list = []

# Create citites with random latitudes and longitudes
lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lng = zip(lat, lng)

# Identify unique cities
for unique_lat_lng in lat_lng:
    city = citipy.nearest_city(unique_lat_lng[0], unique_lat_lng[1]).city_name
    
    
    # Add unique cities to the list
    if city not in city_list:
        city_list.append(city)

len(city_list)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
# Create list to add the variables
city_id= []
city_name = []
country = []
latitude = []
longitude = []
date=[]
humidity = []
temperature = []
wind_speed = []
cloudiness = []
count=0

#Extract weather for various cities from the city list
for city in city_list:
    query_url = f"{url}appid={api_key}&q={city}&units={units}"
    response = requests.get(query_url + city).json()
    try:
        cityid=response["id"]
        cityname=response["name"]
        countrycode=response["sys"]["country"]
        lat=response["coord"]["lat"]
        long=response["coord"]["lon"]
        weatherdate =(datetime.fromtimestamp(response['dt'])).strftime('%m/%d/%y')
        hum=response["main"]["humidity"]
        temp=response["main"]["temp_max"]
        windspeed=response["wind"]["speed"]
        cloud=response["clouds"]["all"]
        
        city_id.append(int(cityid))
        city_name.append(cityname)
        country.append(countrycode)
        latitude.append(lat)
        longitude.append(long)
        date.append(weatherdate)
        humidity.append(hum)
        temperature.append(temp)
        wind_speed.append(windspeed)
        cloudiness.append(cloud)
        
        #Set a count of records found
        count=count+1
        
        print(f"Record {count},city name:{cityname} with city ID: {cityid} was found")
    except:
        print(f"{city}, was not found")
        pass
        
    
```

```python
#Create a DataFrame to hold the weather information
weather_data_df=pd.DataFrame({"City ID":city_id, 
                              "City Name":city_name, 
                              "Country Code":country,
                              "Date": date,
                              "Latitude":latitude,
                              "Longitude":longitude, 
                              "Humidity %":humidity, 
                              "Temperature (F)":temperature,
                              "Wind Speed (mph)":wind_speed, 
                              "Cloudiness %":cloudiness})

#Convert temperature from Kelvin to Fahrenheit
weather_data_df["Temperature (F)"]=(weather_data_df["Temperature (F)"]-273.15)*9/5+32

weather_data_df
```

```python
#Create a file with all the weather information (csv)
weather_data_df.to_csv("weather_data.csv",encoding="utf-8", index=False)
```

## Scatter Plots

Observations:

The City Latitude & Temperature scatter plot makes a good representation of the earth's weather in general. Cities with higher tempetures cluster around the equator, between 0 and 20 degrees, and the tempeture drops as cities get closer to the earth's north and south poles around -40 and around 60 degrees.

## Northern and Southern Hemisphere Plots

```python
#Set two seperate list with northern hemisphere and southern hemisphere data
north_hem=weather_data_df.loc[weather_data_df["Latitude"] >= 0,:]
south_hem=weather_data_df.loc[weather_data_df["Latitude"] < 0,:]

south_hem.head()
```

```python
#Plot regression line on scatter plots
def plot_regression(x_values,y_values,x_axis_label,y_axis_label, hemisphere):
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    regress_values = x_values * slope + intercept
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.title(f" {hemisphere}-City {x_axis_label} & {y_axis_label}")
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    print(f"The r-squared is: {rvalue}")
    print(line_eq)
    plt.show()
```

```python
#plot regression line for northern hemisphere and temperature
plot_regression(north_hem["Latitude"],north_hem["Temperature (F)"],"Latitude","Temperature (F)", "Northern Hemisphere")

plt.savefig('north_temp.png')
```

```python
#plot regression line for southern hemisphere and temperature
plot_regression(south_hem["Latitude"],south_hem["Temperature (F)"],"Latitude","Temperature (F)", "Southern Hemisphere")

plt.savefig('south_temp.png')
```
